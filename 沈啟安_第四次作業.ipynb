## 挑戰題二：身份證字號驗證機
題目說明：我們的身分證是由一個英文字母，外加9個數字組成。而第9個數字其實是檢查碼，也就是說英文字母和前8個數字，會決定第9個數字是什麼。規則參考：https://web.fg.tp.edu.tw/~anny/idtest.htm。

```python
% matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact_manual
```

```python
def check (x):
  num=0
  a=x[0]
  if (a=="A"):
    num=1;
  elif (a=="B"):
    num=10;
  elif (a=="C"):
    num=19;
  elif (a=="D"):
    num=28;
  elif (a=="E"):
    num=37;
  elif (a=="F"):
    num=46;
  elif (a=="G"):
    num=55;
  elif (a=="H"):
    num=64;
  elif (a=="I"):
    num=39;
  elif (a=="J"):
    num=73;
  elif (a=="K"):
    num=82;
  elif (a=="L"):
    num=2;
  elif (a=="M"):
    num=11;
  elif (a=="N"):
    num=20;
  elif (a=="O"):
    num=48;
  elif (a=="P"):
    num=29;
  elif (a=="Q"):
    num=38;
  elif (a=="R"):
    num=47;
  elif (a=="S"):
    num=56;
  elif (a=="T"):
    num=65;
  elif (a=="U"):
    num=74;
  elif (a=="V"):
    num=83;
  elif (a=="W"):
    num=21;
  elif (a=="X"):
    num=3;
  elif (a=="Y"):
    num=12;
  elif (a=="Z"):
    num=30;
  j=8
  for i in range (1,9):
    num+=eval(x[i])*j
    j-=1
  num+=eval(x[9])
  num%=10
  if num==0:
    print("身分證驗證通過")
  else:
    print("身分證驗證不通過")
```

```python
interact_manual(check, x="請輸入身分證")
```
