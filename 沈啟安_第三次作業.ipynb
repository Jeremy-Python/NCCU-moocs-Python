## 挑戰題一：計算微積分期末成績
題目說明：請定義一個全班微積分成績的函式，輸入是一個(50, 3)的array，表示班級中50位學生的成績(小考，期中考，期末考)。
目標輸出則是一個(50, 1)的array，表示所有學生的期末成績，期末成績計算方式如下：

1. 小考20%，期中考35%，期末考45%  
2. 期中考40%，期末考60%。  

擇優作為期末成績。

```python
% matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
#挑戰題一
#隨機產生分數
q1=np.random.randint(0,100,[50,3])
#計算兩種加權方式
s=[]
for i in range (0,50,1):
    s1=q1[i][0]*0.2+q1[i][1]*0.35+q1[i][2]*0.45
    s2=q1[i][0]*0+q1[i][1]*0.4+q1[i][2]*0.6
    if s1>s2:
        s.append(s1)
    elif s1<s2:
        s.append(s2)
#把成績轉成array
fs=np.array(s)
#輸出
print(fs)
```

```python
#第二題
c1=np.random.randint(0,100,[10,5])
c2=np.random.randint(0,100,[10,5])
c3=np.random.randint(0,100,[10,5])
weights=[0.25, 0.25, 0.25, 0.15, 0.10]
```

```python
print(c1*weights)
```

```python
print("class1加權成績:",np.dot(c1,weights))
print("class1加權平均:",np.mean(np.dot(c1,weights)))
print("class2加權成績:",np.dot(c2,weights))
print("class2加權平均:",np.mean(np.dot(c2,weights)))
print("class3加權成績:",np.dot(c3,weights))
print("class3加權平均:",np.mean(np.dot(c3,weights)))
print("class1-3加權平均array:",np.array([np.mean(np.dot(c1,weights)),np.mean(np.dot(c2,weights)),np.mean(np.dot(c3,weights))]))
```
